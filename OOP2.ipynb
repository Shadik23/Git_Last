```python
class Human(object):
  def __init__(self, name, surname, age, city):
    self._name = name
    self._surname = surname
    self._age = age
    self._city = city

def display(self):
  print(f'Имя: {self._name}, Фамилия: {self._surname}, Возраст{self._age}, Город: {self._city} ' )
```

```python
class Student(Human):

  def __init__(self, name, surname, number, evaluations):
    self._number = number
    self._evaluations = evaluations

    Human.__init__(self, name, surname)

  def display(self):
    print(f'Имя: {self._name}, Фамилия: {self._surname}')
    print(f'Класс: {self._number}, Оценки: {self._evaluations}')
```

```python
class Device(object):
  def __init__(self):
    self.__view = ''
    self.__storage = ''
    self.__year = ''

  def set_view(self, view):
    self.__view = view

  def get_view(self):
    return self.__view

  def set_storage(self, storage):
    self.__storage = storage

  def get_storage(self):
    return self.__storage

  def set_year(self, year):
    self.__year = year

  def get_year(self):
    return self.__year

  def info(self):
    print('Вид устройство: ', self.__view, 'Память: ', self.__storage, 'Год: ', self.__year)

```

```python
class Phone(Device):

  def __init__(self):

    super().__init__()
    self.__mark = ''
    self.__model = ''

    def set_mark(self, mark):
       self.__mark = mark

    def get_mark(self):
       return self.__mark

    def set_model(self, model):
      self.__model = model

    def get_model(self):
      return self.__model

    def info(self):

      super().info()
      print('Марка: ', self.__mark, 'Модель: ', self.__model)
```

```python
mob = Phone()
mob.set_view('Сенсорный')
mob.set_storage('128 гб')
mob.set_year('2016')
mob.set_mark('Iphone')
mob.set_model('8 Pro')

mob.info()
```
